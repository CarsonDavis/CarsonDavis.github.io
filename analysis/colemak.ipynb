```python
monkey_colemak_df = monkeydf[monkeydf['tags'].str.contains('6549a314c46e6fc4875bdfdf')]
monkey_colemak_df['date'] =pd.to_datetime(monkey_colemak_df['timestamp'],unit='ms')
monkey_colemak_df = monkey_colemak_df.groupby(monkey_colemak_df['date'].dt.date)['testDuration'].sum().reset_index()
monkey_colemak_df['testDuration_min'] = monkey_colemak_df['testDuration']/60
monkey_colemak_df['date'] = pd.to_datetime(monkey_colemak_df['date'])
monkey_colemak_df
```

```python
carson_colemak=df['carson-colemak_club']

# Add a new column for total time in decimal minutes
carson_colemak['total_time_min'] = carson_colemak['min'] + carson_colemak['sec'] / 60

# Group by date and sum the total time
grouped_df = carson_colemak.groupby('date').agg({'total_time_min': 'sum'}).reset_index()

# Display the resulting DataFrame
carson_colemak =grouped_df
```

```python
carson_ngram=df['carson-ngram']

# Add a new column for total time in decimal minutes
carson_ngram['total_time_min'] = carson_ngram['min'] + carson_ngram['sec'] / 60

# Group by date and sum the total time
grouped_df = carson_ngram.groupby('date').agg({'total_time_min': 'sum'}).reset_index()

# Display the resulting DataFrame
carson_ngram = grouped_df
```

```python
carson_colemak.rename(columns={'total_time_min': 'total_time_min_colemak'}, inplace=True)
carson_ngram.rename(columns={'total_time_min': 'total_time_min_ngram'}, inplace=True)

# Merge the DataFrames on the 'date' column
combined_df = pd.merge(carson_colemak, carson_ngram, on='date', how='outer').fillna(0)

combined_df['date'] = pd.to_datetime(combined_df['date'])
combined_df = pd.merge( monkey_colemak_df, combined_df, on='date', how='outer').fillna(0)

# Display the combined DataFrame
print(combined_df)
```

```python
combined_df
```

```python
# Create the stacked area chart
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=combined_df['date'],
    y=combined_df['total_time_min_colemak'],
    mode='lines',
    stackgroup='one',  # Define stack group
    name='Colemak Club'
))

fig.add_trace(go.Scatter(
    x=combined_df['date'],
    y=combined_df['total_time_min_ngram'],
    mode='lines',
    stackgroup='one',  # Define stack group
    name='Ngrams'
))

fig.add_trace(go.Scatter(
    x=combined_df['date'],
    y=combined_df['testDuration_min'],
    mode='lines',
    stackgroup='one',  # Define stack group
    name='MonkeyType'
))

# Customize the layout
fig.update_layout(
    title='Time Spent Learning Colemak',
    xaxis_title='Date',
    yaxis_title='Minutes per Day',
    template='plotly_dark',
    plot_bgcolor='rgba(0,0,0,0)',
    paper_bgcolor='rgba(0,0,0,0)',
)

# Show the figure
fig.show()

fig_html = pio.to_html(fig, full_html=False)

with open('../_includes/embedded_plotly_graph.html', 'w', encoding='utf-8') as file:
    file.write(fig_html)
```
